## Converting wav files to spectograms

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wavfile

def plot_spectogram(wav_file_path):
    Fs, aud = wavfile.read(wav_file_path)
    # select left channel only
    aud = aud[:,0]
    powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(aud, Fs=Fs)
```

```python
plot_spectogram("audio/hello_shortened.wav")
```

```python
plot_spectogram("audio/goodbye.wav")
```

```python
plot_spectogram("audio/light_on.wav")
```

```python
plot_spectogram("audio/light_off.wav")
```
